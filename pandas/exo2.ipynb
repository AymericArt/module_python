```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sales_predictions = pd.read_csv("data/sales_predictions.csv")
```

```python
sb.relplot(data=sales_predictions, x="item_price", y="date")
```

```python
sales_predictions.head()
```

```python
d = sb.relplot(data=sales_predictions.sample(50), x="item_price", y="date")
d.fig.set_figwidth(15.27)
d.fig.set_figheight(6.7)
```

```python
sales_predictions['date'] = pd.to_datetime(sales_predictions['date'])
```

```python
d = sb.relplot(data=sales_predictions.sample(50), x="item_price", y="date", hue="item_price")
d.fig.set_figwidth(15.27)
d.fig.set_figheight(6.7)
```
